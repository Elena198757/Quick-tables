```python


# Изменение размеров маски для соответствия размеру изображения
resized_mask = cv2.resize(blue_mask, (image5.shape[1], image5.shape[0]))

# Применение маски к изображению
result_mask = cv2.bitwise_and(image5, image5, mask=~resized_mask)

```

```python
# Распознавание текста на изображении
result7 = reader.readtext(result_mask)

# Вывод результатов
for detection in result7:
    text7 = detection[0]
    confidence7 = detection[1]
    print(f"Текст: {text7}, Уверенность: {confidence7}")
```

```python

# Преобразование в цветовое пространство HSV
hsv_image1 = cv2.cvtColor(image5, cv2.COLOR_BGR2HSV)

# Определение диапазона цвета синего
blue_color = np.uint8([[[0, 0, 255]]])  # Синий цвет в BGR
hsv_blue = cv2.cvtColor(blue_color, cv2.COLOR_BGR2HSV)

lower_blue = np.array([hsv_blue[0][0][0] - 10, 50, 50])  # Нижний предел цвета синего в HSV
upper_blue = np.array([hsv_blue[0][0][0] + 10, 255, 255])  # Верхний предел цвета синего в HSV

# Создание маски
blue_mask1 = cv2.inRange(hsv_image1, lower_blue, upper_blue)

# Применение маски к изображению
result8 = cv2.bitwise_and(image5, image5, mask=~blue_mask1)


```

**Пробую заменить синие буквы на белы**

```python
hsv_image2 = cv2.cvtColor(image5, cv2.COLOR_BGR2HSV)
```

```python
lower_blue = (90, 50, 50)  # Нижний предел цвета синего в HSV
upper_blue = (130, 255, 255)  # Верхний предел цвета синего в HSV

blue_mask2 = cv2.inRange(hsv_image2, lower_blue, upper_blue)
```

```python
blue_areas = cv2.bitwise_and(image5, image5, mask=blue_mask2)
```

```python
non_blue_mask = cv2.bitwise_not(blue_mask2)
```

```python
white_text = cv2.bitwise_or(blue_areas, cv2.cvtColor(non_blue_mask, cv2.COLOR_GRAY2BGR))
```

```python
plt.figure(figsize=(20, 20)) 
plt.imshow(cv2.cvtColor(white_text, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```

```python

# Преобразование в цветовое пространство HSV
hsv_image3 = cv2.cvtColor(image5, cv2.COLOR_BGR2HSV)

# Определение диапазона цвета синего
lower_blue = np.array([90, 50, 50])  # Нижний предел цвета синего в HSV
upper_blue = np.array([130, 255, 255])  # Верхний предел цвета синего в HSV

# Создание маски для синего цвета
blue_mask3 = cv2.inRange(hsv_image3, lower_blue, upper_blue)

# Применение инверсии маски для синего цвета
non_blue_mask3 = cv2.bitwise_not(blue_mask3)

# Замена синего цвета на белый
hsv_image3[np.where(blue_mask3 == 255)] = [0, 0, 255]  # Белый цвет в HSV

# Применение инверсии маски для всех остальных цветов
hsv_image3[np.where(non_blue_mask3 == 255)] = [0, 0, 0]  # Черный цвет в HSV

# Преобразование обратно в цветовое пространство BGR
result_image3 = cv2.cvtColor(hsv_image3, cv2.COLOR_HSV2BGR)


```

```python
plt.figure(figsize=(20, 20)) 
plt.imshow(cv2.cvtColor(result_image3, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
