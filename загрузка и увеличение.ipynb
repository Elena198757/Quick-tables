```python
from PIL import Image
import os
import matplotlib.pyplot as plt

folder_path = 'foto'

# Получение списка файлов в папке
file_list = os.listdir(folder_path)

# Создание сетки для отображения изображений
num_cols = 4
num_rows = (len(file_list) - 1) // num_cols + 1

# Создание фигуры и осей для сетки изображений
fig, axes = plt.subplots(num_rows, num_cols, figsize=(12, 12))

# Отображение каждого изображения
for i, file_name in enumerate(file_list):
    if file_name.endswith('.jpg') or file_name.endswith('.png'):
        image_path = os.path.join(folder_path, file_name)
        img = Image.open(image_path)

        # Определение позиции текущего изображения в сетке
        row_idx = i // num_cols
        col_idx = i % num_cols

        # Отображение изображения на соответствующей позиции в сетке
        axes[row_idx, col_idx].imshow(img)
        axes[row_idx, col_idx].axis('off')

# Показ сетки изображений
plt.tight_layout()
plt.show()
```

```python
import os
import cv2

folder_path = 'foto'

# Получение списка файлов в папке
file_list = os.listdir(folder_path)

# Перебор каждого файла в папке
for file_name in file_list:
    if file_name.endswith('.jpg') or file_name.endswith('.png'):
        image_path = os.path.join(folder_path, file_name)
        
        # Загрузка изображения
        image = cv2.imread(image_path)
        
        # Увеличение размера изображения
        scale_percent = 150  # Процент увеличения (в данном случае на 50%)
        width = int(image.shape[1] * scale_percent / 100)
        height = int(image.shape[0] * scale_percent / 100)
        resized_image = cv2.resize(image, (width, height))
        
        # Сохранение увеличенного изображения
        new_image_path = os.path.join("resized_foto", 'resized_' + file_name)
        cv2.imwrite(new_image_path, resized_image)
```
